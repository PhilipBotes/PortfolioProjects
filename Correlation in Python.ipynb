```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Read in the data

df = pd.read_csv(r'C:\Users\Peter\Desktop\Philip\Besigheid\Data Analyst\Portfolio Projects\4. Correlation in Python\Dataset\movies.csv')
```

```python
#Looking at the data

df.head()
```

```python
# Looping through the data to see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
```

```python
# Data Types of Columns

print(df.dtypes)
```

```python
# Change data type of  columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Creat correct year column

df['yearcorrect'] = df['released'].astype(str).str[:4]
df.head()
```

```python
# Ordering data by gross revenue

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
# How to change display options

pd.set_option('display.max_rows', None)
```

```python
# Shows unique company values

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show()
```

```python
# Plott budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'yellow'})
```

```python
# Looking at correlations

df.corr() # Pearson
```

```python
# Showing correlations as heatmap

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# How to look at correlations with non-numeric data

df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
# Stacking correlations

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
# Sorting stacked correlations

sorted_pairs = corr_pairs.sort_values()

sorted_pairs.tail(25)
```

```python
#Identifying high correlations

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
